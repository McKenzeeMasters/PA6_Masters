```python
import math
class Mc_Calc():
    def __init__(self,input1,input2,input3):
        self.input1=input1
        self.input2=input2
        self.input3=input3
    
    def add(num1,num2):
        print(f"Addition = {num1+num2}")
        
    def sub(num1,num2):
        print(f"Subtraction = {num1-num2}")
        
    def multi(num1,num2):
        print(f"Multiplication = {num1*num2}")
        
    def divi(num1,num2):
        print(f"Division = {num1/num2}")
        
    def percent(num1):
        print(f"Percentage = {num1*100}%")
        
    def per_dec(num1):
        print(f"Decimal = {num1/100}")
    
    def Exponent(num1, num2):
        print(f"Exponent = {num1**num2}")
        
    def square(num1):
        print(f"{num1} Squared = {num1**2}")
        
    def SquareRoot(num1):
        print(f"Square Root of {num1} = {num1**0.5}")
    
    def Root(num1,num2):
        print(f"{num2} Root of {num1} = {num1**(1/num2)}")
    
    def Area_Sqr_Rec(num1,num2):
        print(f"Area of the square/rectangle = {num1*num2}")
        
    def Area_Tri(num1,num2):
        print(f"Area of the triangle = {1/2*num1*num2}")
        
    def Area_Cir(num1):
        print(f"Area of the circle = {math.pi*(num1**2)}")
        
    def Pyth(num1,num2):
        print(f"Side c = {(((num1**2)+(num2**2)))**0.5}")
        
    def Area_Trap(num1, num2, num3):
        print(f"Area of the trapezoid = {(0.5*((num1+num2)*num3))}")
        
while True:
    print("Chose from the following menu: ")
    print("1) Addition")
    print("2) Subtraction")
    print("3) Multiplication")
    print("4) Division")
    print("5) Square")
    print("6) Exponent")
    print("7) Square Root")
    print("8) Root")
    print("9) Decimals to Percentages")
    print("10) Percentages to Decimals")
    print("11) Area of a Trapezoid")
    print("12) Area of a Square/Rectangle")
    print("13) Area of a Triangle")
    print("14) Area of a Circle")
    print("15) Pythagorean Theorem")
    print("16) Exit")
    
    choice=int(input("Enter choice (1-16): "))
    
    if choice==1:
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
        Mc_Calc.add(num1,num2)
    elif choice==2:
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
        Mc_Calc.sub(num1,num2)
    elif choice==3:
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
        Mc_Calc.multi(num1,num2)
    elif choice==4:
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
        if num2==0:
            print("Infinity")
        else:
            Mc_Calc.divi(num1,num2)
    elif choice==5:
        num1=float(input("Enter base number: "))
        Mc_Calc.square(num1)
    elif choice==6:
        num1=float(input("Enter base number: "))
        num2=float(input("Enter exponent: "))
        if num2<0:
            print("Error")
        else:
            Mc_Calc.Exponent(num1,num2)
    elif choice==7:
        num1=float(input("Enter base number: "))
        if num1<0:
            print("Error")
        else:
            Mc_Calc.SquareRoot(num1)
    elif choice==8:
        num1=float(input("Enter base number: "))
        num2=float(input("Enter root number: "))
        if num1<0:
            print("Error")
        else:
            Mc_Calc.Root(num1,num2)
    elif choice==9:
        num1=float(input("Enter decimal: "))
        Mc_Calc.percent(num1)
    elif choice==10:
        num1=float(input("Enter percentage:  "))
        Mc_Calc.per_dec(num1)
    elif choice==11:
        num1=float(input("Enter base 1: "))
        num2=float(input("Enter base 2: "))
        num3=float(input("Enter height: "))
        Mc_Calc.Area_Trap(num1, num2, num3)
    elif choice==12:
        num1=float(input("Enter side one: "))
        num2=float(input("Enter side two: "))
        Mc_Calc.Area_Sqr_Rec(num1,num2)
    elif choice==13:
        num1=float(input("Enter the base: "))
        num2=float(input("Enter the height: "))
        Mc_Calc.Area_Tri(num1,num2)
    elif choice==14:
        num1=float(input("Enter the radius: "))
        Mc_Calc.Area_Cir(num1)
    elif choice==15:
        num1=float(input("Enter side a: "))
        num2=float(input("Enter side b: "))
        Mc_Calc.Pyth(num1,num2)
    elif choice==16:
        break
    else: 
        print("Invalid choice")
```
